# matplotlib의 2가지 코딩 스타일

## OO-스타일: matplotlib의 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot 스타일보다 복잡함

```python
x = np.linspace(0, 2, 100)
x 
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7)) #튜플 형태로 줘야함
ax.plot(x, x, label='linear') #x,x끼리 같은 숫자 준 거
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3,1, figsize=(5, 7))
ax[0].plot(x, x, label='linear') 
ax[1].plot(x, x**2, label='quadratic')
ax[2].plot(x, x**3, label='cubic') 
ax[0].set_xlabel('x label')
ax[0].set_ylabel('y label')
ax[0].set_title('간단 그래프')
ax[1].set_xlabel('x label')
ax[1].set_ylabel('y label')
ax[1].set_title('quadratic 간단 그래프')
ax[2].set_xlabel('x label')
ax[2].set_ylabel('y label')
ax[2].set_title('cubic 간단 그래프')
plt.tight_layout()
plt.show()
```

# pyplot 스타일
* OO-style보다 사용법 간단
* 복잡한 그래프를 그릴 때는 적합하지 않다.

```python
plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label='linear') 
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic') 
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('pyplot 간단 그래프')
plt.legend()
plt.show()
```

# 선 모양 바꾸기 set_linestyle(모양)

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python
data1
```

```python
data2
```

```python
# 선모양: '-', '--', '-.', ':'
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle=":")
ax.plot(x, np.cumsum(data2), color='orange', linewidth=2, linestyle="-.")
plt.show()
```

# 눈금자 단위 구간 설정하기

```python
xdata = np.arange(len(data1))
fig, ax = plt.subplots(2, 1, figsize=(5, 2.7), layout='constrained')
ax[0].plot(xdata, data1)
ax[0].set_title("automatic ticks")

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0,100,30), ['zero', '30', 'sixty', '90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title("manual ticks")
plt.show()
```
