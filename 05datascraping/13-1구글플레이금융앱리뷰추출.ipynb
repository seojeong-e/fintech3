### 국민은행

```python
import time
import pandas as pd
from datetime import datetime,timedelta
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
#웹브라우저를 실행 할 때 사용할 옵션 - 사람인 것처럼 정보 입력
options = Options()
options.add_experimental_option("detach", True)
options.add_argument("start-maximized")
options.add_argument("Chrome/141.0.0.0")
options.add_argument("lang=ko_KR")


driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
    )
#웹브라우저에서 url 열기
driver.get("https://play.google.com/store/apps/details?id=com.kbstar.kbbank&hl=ko")

driver.execute_script(f"window.scrollTo(0,1400)")
time.sleep(2)

#평점 및 리뷰 옆의 -> 버튼 클릭
kb_button = driver.find_element(By.CSS_SELECTOR, 'button[aria-label*="평점 및 리뷰 자세히 알아보기"]')
kb_button.click()

# 리뷰를 정렬하기 위해서 버튼 클릭
driver.find_element(By.ID, "sortBy_1").click()
time.sleep(2)

# '최신'을 찾아서 클릭
driver.find_element(By.CSS_SELECTOR, 'span[aria-label*="최신"]').click()
time.sleep(2)

today = datetime.today()
one_month_ago = today - timedelta(days=7)

# 리뷰가 담긴 창을 찾아서 Javascript로 1000px씩 아래로 스크롤
driver.execute_script("document.querySelector('.fysCi.Vk3ZVd').scrollBy(0,1000)")

# 리뷰 목록 가져오기
review_list =driver.find_elements(By.CSS_SELECTOR, ".RHo1pe")
result = {}
cols = ('리뷰일','별점','사용자 리뷰','회사 응답')
for review in review_list:

    # 리뷰일
    kb_review_date = driver.find_element(By.CSS_SELECTOR,'.bp9Aid').get_attribute('innerHTML')

    # 추출한 리뷰일을 날짜형 데이터로 변경
    kb_review_date = kb_review_date.replace(" ","").replace("년","-").replace("월","-").replace("일","")
    kb_review_date = datetime.strptime(kb_review_date, "%Y-%m-%d")

    # 별점
    kb_rating = float(driver.find_element(By.CSS_SELECTOR, 'div[aria-label*="별표 5개 만점에"]').get_attribute('aria-label').split()[3].replace("개를",""))

    # 사용자 리뷰(텍스트)
    kbuser_review = driver.find_element(By.CSS_SELECTOR, ".h3YV2d").get_attribute('innerHTML')

    # 회사 응답
    kbcompany_reply = driver.find_element(By.CSS_SELECTOR,".ras4vb").text.replace("\n","")
    
    values = (kb_review_date, kb_rating, kbuser_review, kbcompany_reply)
    for key, value in zip(cols, values):
        result.setdefault(key, []).append(value)
        
df = pd.DataFrame(result)
df
```
