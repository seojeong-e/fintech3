* num_maker - 정방향, 역방향 리스트가 출력되는 함수
* 3개의 숫자를 입력받아서 숫자 리스트를 만들고, 오름차순 정렬 반환
* 만들어진 리스트에 + 3을 한 결과를 리스트로 만들고, 내림차순 정렬 반환
* range(a, b, c) (시작, 끝, 스텝)
* return : list_result, reversed_list

```python
def num_maker(start, end, step):
    list_result = list(range(start, end+1, step))
    reversed_result = [] #빈 리스트 만들기
    for i in list_result:
        reversed_result.append(i+3)
    return list_result, reversed_result[::-1] #역순 출력
```

```python
num_maker(1, 10, 1)
```

```python
def num_maker(start, end, step):
    list_result = list(range(start, end+1, step))
    reversed_result = [i+3 for i in list_result] 
    return list_result, reversed_result[::-1] 
```

```python
num_maker(1, 10, 1)
```

# 위치인수와 가변인수 동시에 받기
* 위치인수와 가변인수를 동시에 받을 때는 반드시 **위치인수**가 먼저 와야 한다.
* def add(a, *args): OK
* def add(*args, a): X

```python
# 정수 1개와 리스트를 입력받아 리스트의 요소에 정수를 곱하는 함수
def muls(a, *args):
    print(a, args)
    result = []
    for num in args:
        result.append(num*a)
    return result
```

```python
muls(2, *h)
```

# 키워드 가변인수
* def함수명(**kwargs)
* 입력받는 자료가 address="안알려줌" 변수명 = 자료 형태인 것 
* 키워드 가변인수는 함수 내에서 dict로 변환되어 사용된다.

```python
def personal_info3(**kwargs):
    print(kwargs, type(kwargs))
    key = list(kwargs.keys())
    values = list(kwargs.values())
    return key, values
```

* 1에서 10까지 수중에서 3의 배수만 문자열로 바꿔서 리스트로 출력하세요

```python
def tostr(x):
    result = []
    for i in x:
        if i % 3 == 0:
            result.append(str(i))
        else:
            result.append(i)
    return result
```

```python
tostr(list(range(1, 11)))
```

```python
x = list(range(1, 11))
result = [str(i) if i % 3 == 0 else i for i in x]
print(result)
```

# 예외 처리방법 3가지
* try-except 구문으로 처리
* if else 조건 표현식 처리
* get, setdefault 같은 함수를 사용해 처리

* 학생들의 이름 리스트를 받아 해당 학생의 성적 출력하기

## try except문을 이용한 예외처리

```python
# 학생정보 
student_scores = dict(student=dict(alice=85, bob=92, mark=78))
student_scores
```

```python
# 성적을 조회할 목록
names = ['alice', 'david', 'mark', 'bob']
```

```python
# 반복문을 이용해서 성적 조회
for name in names:
    try:
        score = student_scores['student'][name]
        print(f"{name}: {score}")
    except Exception as e:
#        print(e)
        print(f"{name}의 점수는 없습니다.")
    finally:
        print(f"{name}의 점수 조회가 완료되었습니다.")
        print()
```

## 조건표현식을 사용한 예외처리

```python
# 반복문을 이용해서 성적 조회
for name in names:
    score = student_scores['student'][name] if name in student_scores['student'].keys() else "키 없음"
    print(f"{name}: {score}")
```

## get 함수를 이용해서 처리하기

```python
# 반복문을 이용해서 성적 조회
for name in names:
    score = student_scores['student'].get(name, f"{name}의 점수는 없습니다.")
    print(f"{name}: {score}")
```
