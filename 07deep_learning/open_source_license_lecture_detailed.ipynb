# 📘 오픈소스 라이선스 강의안 (자세 설명 버전)

이 노트북은 오픈소스 라이선스의 개념과 종류, 그리고 실제 사용 방법을 설명하는 **강의용 자료**입니다.


## 🎯 학습 목표

이 강의가 끝나면 수강생은 다음을 스스로 설명할 수 있어야 합니다.

1. **저작권과 오픈소스의 기본 개념**  
   - “오픈소스 = 공짜”라는 오해를 버리고,  
     왜 ‘저작권을 포기하는 것’이 아니라 ‘조건부 허용’인지 이해한다.

2. **주요 오픈소스 라이선스의 차이점**  
   - MIT / Apache / BSD / GPL / LGPL / AGPL / CCL 등의 이름을 들었을 때  
     “대략 어떤 특징인지” 말할 수 있다.

3. **실무에서 라이선스를 확인하고 사용하는 방법**  
   - 프로젝트에서 어떤 오픈소스를 썼는지 파악하고,  
   - 라이선스 의무사항(고지문, 소스 공개 등)을 어떻게 지켜야 하는지 이해한다.


---

# 1️⃣ 저작권과 오픈소스 라이선스


## 1. 오픈소스란 무엇인가?

### 1) 기본 정의

- 어떤 개발자가 소프트웨어를 만들면, 그 소프트웨어의 **저작권은 기본적으로 그 개발자에게** 있습니다.  
- 이 개발자가 “아무나 가져다 써도 된다, 다만 이런 조건들은 꼭 지켜라”라고 **사용 조건을 적어놓은 문서**가 바로 **라이선스**입니다.
- 오픈소스는 그 라이선스가 **소스코드를 공개하고, 사용·수정·재배포를 허용하는 방식으로 설계된 소프트웨어**를 말합니다.

### 2) “Free”는 공짜가 아니다

- FOSS(Free and Open Source Software)에서 **Free는 ‘자유’**를 의미합니다.  
  - “자유롭게 실행할 자유”  
  - “자유롭게 수정할 자유”  
  - “자유롭게 공유할 자유”  
- **가격이 0원**이라는 의미가 아니며, 오픈소스도 **유료로 제공**될 수 있습니다.  
  예) 기업이 오픈소스를 기반으로 부가 기능을 붙여서 유료로 판매.

### 3) 저작권 포기가 아니다

- 오픈소스라고 해서 저작권을 포기한 것이 아닙니다.  
- 오히려 “저작권이 있으니, 이 조건들을 지키면 사용을 허용하겠다”라는 선언에 더 가깝습니다.
- 따라서 **라이선스를 지키지 않으면 저작권 침해**가 될 수 있습니다.


## 2. 오픈소스의 특징

### 1) 소스코드 공개

- 오픈소스는 **소스코드 형태로 배포**되기 때문에:
  - 내부 구현을 배울 수 있고,
  - 버그를 수정하거나 기능을 추가해서 **내 버전으로 확장**할 수 있습니다.
- 이에 비해 상용 소프트웨어는 보통 **실행 파일(.exe, .dll, .so 등)**만 제공되고  
  소스는 공개되지 않으므로 수정이 거의 불가능합니다.

### 2) 협업과 진화

- 누구나 버그를 리포트하고, 코드를 기여할 수 있으므로  
  **짧은 기간에 빠르게 발전**하는 경향이 있습니다.
- 예: Linux 커널, Python, TensorFlow, Kubernetes 등


## 3. 오픈소스의 법적 책임

### 1) 라이선스 준수 의무

- 상용 SW와 마찬가지로 오픈소스를 사용하려면  
  **반드시 해당 라이선스를 지켜야 합니다.**
- 대표적인 의무사항 예:
  - 저작권자 이름/라이선스 종류를 제품에 표기
  - 소스코드 공개 (GPL 계열)
  - 재배포 시 동일 라이선스 적용

### 2) 위반 시 발생 가능한 문제

- 오픈소스 커뮤니티 또는 저작권자가 **법적 대응 또는 클레임**을 제기할 수 있습니다.
- 이미 출하된 제품의 판매를 중단하거나,  
  긴급 패치/오픈소스 공개 등의 조치를 취해야 할 수 있습니다.
- “라이선스를 제대로 관리하지 않는 회사”라는 이미지가 생길 수도 있습니다.


## 4. 오픈소스 라이선스란?

### 1) 정의

- **저작권자와 사용자 사이의 계약서**라고 생각하면 됩니다.
- 이 문서에는:
  - 사용자가 할 수 있는 일 (허용 사항)
  - 반드시 지켜야 하는 조건 (의무사항)
  - 책임의 범위(책임 제한, 보증 부인)  
  등이 자세히 적혀 있습니다.

### 2) 직접 만들 수도 있지만…

- 이론적으로는 저작권자가 자기만의 라이선스를 직접 써도 됩니다.
- 하지만 실제로는 **이미 널리 쓰이는 표준 라이선스(MIT, GPL, Apache 등)**를 선택하는 것이 일반적입니다.  
  - 이유:
    - 법적 검증이 어느 정도 되어 있고  
    - 사용자들이 이해하기 쉽고  
    - 기업에서 정책으로 지원하기도 쉬움


## 5. 오픈소스 라이선스의 핵심 의무사항

### (1) 저작권 표시 및 라이선스 고지

- 거의 모든 라이선스가 요구하는 기본 의무입니다.
- 보통 다음을 포함합니다:
  - “이 제품은 어떤 오픈소스를 사용하고 있으며…”
  - 각 오픈소스의 이름, 저작권자, 라이선스 이름
  - 라이선스 전문 또는 링크
- 책에서 논문 참고문헌을 적는 것과 비슷한 개념입니다.

### (2) 소스코드 공개 (특히 GPL 계열)

- GPL, LGPL, AGPL과 같은 Copyleft 라이선스는  
  **바이너리를 배포할 때 해당 바이너리에 대응하는 소스코드를 함께 제공**할 것을 요구합니다.
- 예:
  - 다운로드 페이지에 “소스코드 다운로드 링크” 제공
  - CD/DVD/USB 등 물리 매체로 제품을 판매할 때,  
    소스코드를 요청하면 제공하겠다는 문구 및 방법 안내

### (3) 재배포 시 동일 라이선스 적용 (Copyleft)

- 소프트웨어를 수정해서 재배포할 경우  
  **원래의 라이선스와 똑같은 라이선스를 적용**해야 하는 경우가 있습니다.
- 대표적으로 **GPL**:
  - GPL 코드를 사용해 만든 프로그램 → **전체를 GPL로 공개해야** 함  
  - 상용/독점 앱에서 사용하기 매우 부담스러운 이유가 여기에 있습니다.

### (4) “배포”가 있을 때만 강하게 발생

- 사내에서만 테스트 목적으로 사용하고 외부로 내보내지 않는다면  
  일반적으로 소스 공개 의무는 발생하지 않습니다.
- 하지만:
  - 클라우드·웹 서비스의 경우 “배포”로 볼 것인지가 라이선스에 따라 다르므로 AGPL 같은 경우 특히 주의가 필요합니다.


## 6. Public Domain

- 어떤 저작물은 **저작권자가 권리를 포기**하거나,  
  **보호기간이 만료**되어 “Public Domain”이 될 수 있습니다.
- 이 경우:
  - 자유롭게 수정, 재배포, 상업적 이용 가능
  - 출처 표기의 의무도 법적으로는 없습니다(그러나 관례상 쓰는 게 좋음).
- 단, **저작자가 불분명하다고 해서 Public Domain은 아닙니다.**
  - 예: 인터넷에서 떠도는 이미지, 코드 조각 → 거의 다 저작권 존재

# 2️⃣ 주요 오픈소스 라이선스


## 1. 전체적인 분포

- 세상에는 수천 개의 오픈소스 라이선스가 있지만,  
  실제로 널리 쓰이는 것은 **MIT, GPL, Apache, BSD, LGPL, MPL, EPL 등 소수**입니다.
- 이 강의에서는 개발자가 실무에서 가장 자주 접하게 되는 것들 위주로 다룹니다.


## 2. Permissive vs Copyleft

### 1) Permissive 라이선스

- 조건이 **“느슨하고” “사용하기 쉬운”** 라이선스.
- 주로 요구하는 것은:
  - 저작권 표시
  - 라이선스 고지
  - 책임 면책

→ 소스코드를 공개하지 않고 상용/독점 소프트웨어에 포함시켜도 되는 경우가 많습니다.

### 2) Copyleft 라이선스

- 조건이 **“강하고” “공유를 강제하는”** 라이선스.
- 대표적인 의무:
  - 파생저작물도 동일 라이선스로 배포
  - 바이너리 배포 시 소스코드 공개

→ 상용 소프트웨어와 결합 시 법적 검토가 필수입니다.


## 3. Permissive 라이선스 자세히

### (1) MIT License

- 가장 단순한 라이선스 중 하나.
- 핵심 내용:
  1. 이 소프트웨어를 사용·복제·수정·병합·발행·배포·재라이선스·판매할 수 있다.  
  2. 단, **위 저작권 표시와 허가 공지(MIT License 문구)를 모든 복제본에 포함**시킬 것.  
  3. 이 소프트웨어는 **“있는 그대로(as-is)” 제공되며, 어떠한 보증도 없다.**
- 실제 예:
  - Node.js, jQuery, Ruby on Rails, VS Code 등 매우 많은 프로젝트가 MIT 사용

### (2) Apache License 2.0

- MIT보다 조금 더 긴 라이선스지만 여전히 Permissive 계열입니다.
- 특징:
  - **특허(Patent) 관련 조항**이 있어, 특허 분쟁 상황에 대비한 구조를 가짐
  - 소스코드를 반드시 공개할 필요는 없지만,
    - 수정한 파일에는 변경 사실을 명시
    - NOTICE 파일 유지 등
- Android, Hadoop 등이 Apache License를 사용합니다.

### (3) BSD License

- Berkeley Software Distribution에서 유래.
- 대표 특징:
  - 소스코드 공개 의무 없음
  - 저작권 표기 및 책임 면책
  - 저작권자 이름을 광고에 임의로 쓰지 말 것(이름 도용 금지)
- Nginx 등 여러 프로젝트에서 사용.


## 4. Copyleft 라이선스 자세히

### (1) GPL (GNU General Public License)

- 가장 “강한” Copyleft 라이선스 중 하나.
- 핵심:
  1. 이 소프트웨어를 사용·복제·수정·배포할 수 있다.  
  2. **배포할 경우** → **소스코드를 함께 제공**해야 한다.  
  3. 이 소프트웨어를 포함한 프로그램을 배포할 경우 → **전체 프로그램을 GPL로 배포**해야 한다.
- 의미:
  - 한 번 GPL을 섞어 버리면, 그 프로그램 전체가 “오픈소스(GPL)”가 되어야 함.
  - 상용/독점 앱에서 사용하기 매우 부담스러운 이유가 여기에 있습니다.

### (2) LGPL (Lesser GPL)

- “라이브러리용 GPL”이라고 이해하면 쉽습니다.
- 목적:
  - 애플리케이션 코드 전체를 GPL로 강제하는 대신,  
    **라이브러리 자체의 자유와 공유를 보장**하려는 타협안.
- 특징:
  - LGPL 라이브러리를 수정해서 배포한다면  
    수정된 라이브러리 부분에 대한 소스를 공개해야 함.
  - 그러나 애플리케이션 전체를 공개할 필요까지는 없을 수 있음(링크 방식 등 세부 규정 존재).

### (3) AGPL (Affero GPL)

- 기존 GPL은 “배포(distribution)”가 일어났을 때만 소스 공개 의무가 명확했습니다.  
  그런데 웹 서비스/클라우드 시대에는:
  - 서버에만 설치하고, 사용자는 웹으로만 접근 → “배포”가 아니라고 주장할 여지가 생김.
- AGPL은 이를 막기 위해:
  - **네트워크를 통해 서비스를 제공하는 것**도 사실상 배포로 간주.
  - 따라서 AGPL 소프트웨어를 서버에서 사용해 웹 서비스를 제공하면  
    **소스코드를 공개해야 하는 의무**가 생깁니다.


## 5. CCL (Creative Commons License)

- 코드가 아니라 **이미지, 사진, 문서, 강의자료** 등에 많이 사용되는 라이선스 체계입니다.
- 보통 네 가지 아이콘이 조합되어 나타납니다:
  - BY (저작자 표시)
  - NC (비영리)
  - ND (변경 금지)
  - SA (동일 조건 변경 허락)
- 예:
  - “CC BY-NC-SA” → 저작자 이름 표기만 하면 사용 가능,  
    다만 비영리로만, 변경했을 경우 동일 조건으로 공유해야함.

# 3️⃣ 오픈소스 사용 방법


## 1. “우리가 어떤 오픈소스를 쓰고 있는지” 파악하기

실무에서 가장 큰 문제는,  
**개발자들은 열심히 가져다 쓰는데, 팀/회사 차원에서 무엇을 썼는지 목록이 없다는 것**입니다.

이를 위해:

### 1) 직접 확인

- 각 오픈소스 패키지의 최상단 디렉터리에 보통 이런 파일이 있습니다.
  - `LICENSE`
  - `COPYING`
  - `NOTICE`
- GitHub, GitLab 등에서는:
  - 저장소 루트에 `LICENSE` 파일이 거의 항상 있습니다.
  - 해당 파일을 읽으면 라이선스 종류와 조건이 나옵니다.

### 2) 도구 사용(SCA: Software Composition Analysis)

- 소스코드를 스캔해서 어떤 오픈소스가 포함되어 있는지 자동으로 검출해 주는 도구들:
  - 문자열 패턴으로 찾는 단순 도구
  - 실제 코드 구조를 비교하는 고급 도구
  - 바이너리만으로도 특정 오픈소스를 추정하는 도구
- 큰 조직/프로젝트에서는 이런 툴을 **CI/CD 파이프라인에 연동**해서  
  빌드 시 자동으로 오픈소스를 분석·리포트 하기도 합니다.


## 2. 라이선스를 못 찾겠다면?

- 코드가 인터넷 어딘가에 올라와 있는데 라이선스 표기가 없다면:
  - “이건 공짜겠지?”라고 생각하면 위험합니다.
  - 최소한 다음을 시도해야 합니다:
    1. 해당 프로젝트 페이지를 다시 꼼꼼히 읽어보기  
    2. 코드 내 주석 탐색  
    3. 그래도 없으면 **작성자에게 이메일/이슈로 문의**
- 많은 오픈소스 개발자들은  
  “이 코드를 사용해도 되나요? 라이선스는 무엇인가요?”라는 질문에  
  친절하게 답해주는 경우가 많습니다.


## 3. 코드 작성 시 실무적인 유의사항

### 1) 고지문(OSS Notice) 작성

- 제품 출하 시 다음 정보가 포함된 문서를 하나 만드는 것을 권장:
  - “이 제품은 다음과 같은 오픈소스 소프트웨어를 사용합니다.”
  - 각 오픈소스의 이름, URL, 저작권자, 라이선스 종류
  - 문의 이메일 주소 등
- 모바일 앱/웹 서비스라면:
  - 설정 → “오픈소스 라이선스” 메뉴에 이 내용을 표시하는 것이 일반적입니다.

### 2) 디렉터리 단위로 사용하기

- 함수 몇 줄만 복붙해서 사용하면:
  - 나중에 “이 코드가 어디에서 왔는지” 추적하기 어려워집니다.
- 가능하면:
  - **오픈소스는 하나의 디렉터리로 묶어서 관리**  
  - 원래 프로젝트 이름을 그대로 유지  
  - `LICENSE`, `README` 파일 함께 포함
- 이렇게 하면 나중에 라이선스 점검/감사 시 매우 큰 도움이 됩니다.

### 3) 외부 코드/데이터는 별도 디렉터리로

- 자사가 저작권을 가지지 않은 코드/데이터(3rd party, 오픈소스, 샘플 데이터 등)는
  - `/external`
  - `/third_party`
  - `/reference`
  같은 폴더 아래로 모아서 관리하는 습관을 들이면 좋습니다.
- 이렇게 해두면:
  - “우리가 직접 작성한 코드”와  
  - “외부에서 가져온 코드”를 명확하게 구분할 수 있습니다.

# 🔚 최종 정리

- 오픈소스는 **저작권을 포기한 무료 소프트웨어가 아니라**,  
  저작권자가 정해 놓은 **조건을 지키는 범위 내에서 자유롭게 사용할 수 있는 소프트웨어**입니다.
- 라이선스는 크게:
  - **Permissive (MIT, BSD, Apache)**  
  - **Copyleft (GPL, LGPL, AGPL)**  
  로 나뉘며, Copyleft는 소스 공개와 동일 라이선스 적용 의무가 강합니다.
- 실무에서 가장 중요한 것은:
  1. 어떤 오픈소스를 사용하고 있는지 목록을 관리하기  
  2. 해당 라이선스의 의무사항(고지문, 소스 공개 등)을 이해하고 지키기  
  3. 애매하거나 표기가 없으면 **저작권자에게 문의**하기
