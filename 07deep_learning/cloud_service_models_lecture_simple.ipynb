# ☁️ 클라우드 컴퓨팅 강의안  
### — Cloud Service Models (IaaS / PaaS / SaaS) & Deployment Models —

이 노트북은 **클라우드 서비스 모델(IaaS / PaaS / SaaS)** 와 **클라우드 배포 모델**을 설명하는 강의용 자료입니다.


## 1️⃣ 클라우드 컴퓨팅이란?

클라우드 컴퓨팅(Cloud Computing)은  
**인터넷을 통해 서버·스토리지·데이터베이스·네트워크·소프트웨어 등을 서비스 형태로 제공**하는 기술입니다.

### ✔ 주요 특징
- **온디맨드(On-Demand)**: 필요할 때 즉시 사용 가능  
- **확장성(Scalability)**: 트래픽 증가 시 자동 확장  
- **비용 효율성(Cost Efficiency)**: 사용한 만큼 비용 지불 (Pay-as-you-go)  
- **관리 부담 감소**: 물리 서버 관리 불필요  


## 2️⃣ 클라우드 서비스 모델 (Cloud Service Models)

클라우드 서비스는 아래 3단계로 구분됩니다.

| 구분 | IaaS | PaaS | SaaS |
|------|------|------|------|
| 풀네임 | Infrastructure as a Service | Platform as a Service | Software as a Service |
| 제공 범위 | 서버·스토리지·네트워크 | 개발/배포 플랫폼 | 완성된 소프트웨어 |
| 사용자는 무엇을 관리? | OS, 미들웨어, 앱 관리 필요 | 앱 개발만 하면 됨 | 사용만 하면 됨 |
| 예시 | AWS EC2, GCP Compute Engine | Heroku, AWS Elastic Beanstalk | Gmail, Slack, Figma |


## 3️⃣ IaaS (Infrastructure as a Service)

### 🧱 정의
가상 서버, 스토리지, 네트워크, 운영체제 같은 **IT 인프라를 서비스 형태로 제공**.

### ✔ 사용자 역할
- OS 설치/업데이트  
- 미들웨어(Web Server, DB 설치)  
- 앱 배포  
- 네트워크 설정  

### ✔ 장점
- 높은 자유도  
- 물리 서버 구매 비용 절감  
- 탄력적 확장(Easy scaling)

### ✔ 단점
- 관리 부담이 크다  
- 보안/운영/모니터링을 사용자가 처리해야 함

### ✔ 대표 서비스
- AWS EC2  
- Google Compute Engine  
- Azure Virtual Machine  
- DigitalOcean Droplet  


## 4️⃣ PaaS (Platform as a Service)

### 🧩 정의
개발자가 **코드 개발과 배포에 집중할 수 있도록 플랫폼 제공**.

### ✔ 제공 항목
- 런타임(Node, Python, Java 등)  
- 애플리케이션 서버  
- 데이터베이스 서버  
- CI/CD 환경  
- 자동 스케일링  

### ✔ 장점
- 서버/OS 관리 필요 없음  
- 자동 확장  
- 개발 속도 향상  

### ✔ 단점
- 환경 제약 (자유도 낮음)  
- 벤더 종속성(Vendor-Lock-in)

### ✔ 대표 서비스
- Heroku  
- Google App Engine  
- AWS Elastic Beanstalk  
- Firebase  
- Render  


## 5️⃣ SaaS (Software as a Service)

### 🖥 정의
소프트웨어를 설치 없이 **웹으로 바로 사용하는 형태**.

### ✔ 특징
- 로그인만 하면 사용 가능  
- 유지보수/업데이트는 공급자가 처리  
- 완성된 서비스 제공  

### ✔ 장점
- 설치 불필요  
- 자동 업데이트  
- 협업 기능 우수  

### ✔ 단점
- 커스터마이징 제한  
- 인터넷 연결 필수  
- 데이터가 외부에 저장됨  

### ✔ 대표 서비스
- Gmail, Outlook  
- Slack, Zoom  
- Notion, Figma  
- Salesforce  


## 6️⃣ IaaS — PaaS — SaaS 비교 도식 (개념)

### 🔍 사용자 책임 범위 (개념 요약)

- **On-Premise** : 모든 것을 직접 관리  
- **IaaS**       : OS 위부터 사용자가 관리  
- **PaaS**       : 앱과 데이터만 관리  
- **SaaS**       : 사용만 하면 됨  

### 시각화 표

| 관리항목 | 온프레미스 | IaaS | PaaS | SaaS |
|---------|------------|------|------|------|
| 애플리케이션 | ✔ 직접 | ✔ | ✔ | ✖ |
| 데이터 | ✔ | ✔ | ✔ | ✖ |
| 런타임 | ✔ | ✔ | ✖ | ✖ |
| 미들웨어 | ✔ | ✔ | ✖ | ✖ |
| OS | ✔ | ✔ | ✖ | ✖ |
| 가상화 | ✔ | ✖ | ✖ | ✖ |
| 서버/스토리지 | ✔ | ✖ | ✖ | ✖ |
| 네트워크 | ✔ | ✖ | ✖ | ✖ |


## 7️⃣ 클라우드 배포 모델 (Deployment Models)

### 1) Public Cloud (공용 클라우드)
- AWS, Azure, GCP
- 불특정 다수가 공동 인프라 사용
- 비용 효율적, 빠른 구축  

### 2) Private Cloud (사설 클라우드)
- 특정 기업 전용 클라우드
- 높은 보안성, 비용 비쌈  
- 예: 금융·공공기관 내부 클라우드

### 3) Hybrid Cloud (하이브리드)
- Public + Private 조합  
- 민감한 데이터는 Private  
- 서비스는 Public 활용  
- 예: 금융 서비스에서 고객 데이터 private, 앱은 public

### 4) Multi Cloud (멀티 클라우드)
- AWS + GCP + Azure 등 여러 클라우드 동시 사용  
- 벤더 종속성 최소화  
- 고가용성 확보  


## 8️⃣ 클라우드 도입 장점

### ✔ 기술적 장점
- 무중단 운영  
- 글로벌 배포 가능  
- 자동 확장(Auto Scaling)  
- 보안 및 백업 체계 제공  

### ✔ 비즈니스 장점
- 초기 비용 감소 (CAPEX → OPEX)  
- 빠른 시장 출시(Time to Market 단축)  
- 유지관리 부담 감소  


## 9️⃣ 실제 기업 사용 예시

### 스타트업
- SaaS 기반 협업 도구 → Notion, Slack  
- PaaS 기반 배포 → Heroku, Firebase  
- IaaS 기반 백엔드 → AWS EC2  

### 대기업
- Hybrid Cloud로 보안 강화  
- Kubernetes(GKE/EKS) 기반 자동 확장  
- 데이터 분석 → BigQuery / Redshift  


## 🔟 정리 (요약)

### 한 줄 정리 🚀
- **IaaS**: 서버(인프라) 빌려 쓰는 것  
- **PaaS**: 개발 환경까지 제공  
- **SaaS**: 소프트웨어 완제품 제공  
- **Deployment Models**: Public / Private / Hybrid / Multi  

이 노트북은 강의안으로 바로 활용할 수 있고,  
필요시 여기에 **과제, 퀴즈, 실제 서비스 스크린샷** 등을 추가해서 확장할 수 있습니다.
